### Numerical integration exercise

Take the variables $C_A$, $C_B$ and $C_C$ where:

$$\frac{dC_i}{dt}=r_i$$

The rates of the 3 components is given by:

$$-r_A=kC_AC_B \qquad\qquad r_B=2kC_AC_B \qquad\qquad r_C=1.5kC_AC_B $$

The rate function is given below:


```python
def ratefun(C):
    Ca, Cb, Cc = C[0],C[1],C[2]   
    
    k=0.1
    
    return [-k*Ca*Cb,
            2*k*Ca*Cb,    
            1.5*k*Ca*Cb]
```

Define a function **intfun** (similar to **ferm-eq** in your notes) that obtains the rates from **ratefun** and returns the differentials to **odeint** 

1) Given the initial value of $[C_A, C_B, C_C]=[2,0.1,0]$ at $t=0$, plot $C_A$, $C_B$ and $C_C$ as a function of time for upto 25 minutes.

2) Determine at what time $C_A=0.006$

3) What will happen if the value of k is varied. Postulate an answer before testing with a graph.

4) Plot $r_B$ as a function of time: 

5) At what moment in time is $r_B$ at a maximum?

```python
from __future__ import division 
from numpy import linspace
from numpy import array
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint

def ratefun(C):
    Ca, Cb, Cc = C[0],C[1],C[2]   
    
    k=0.1
    
    return [-k*Ca*Cb,
            2*k*Ca*Cb,    
            1.5*k*Ca*Cb]

def intfun(C,t):
    Ca, Cb, Cc = C[0], C[1], C[2]
    r = ratefun([Ca, Cb, Cc])
    return r[0], r[1], r[2]

c0 = [2,0.1,0]
tspan = linspace(0,25,200)
C = odeint(intfun,c0,tspan)
C1 = C[:,0]
C2 = C[:,1]
C3 = C[:,2]

plt.plot(tspan, C1, color='blue', label = 'C1')
plt.plot(tspan, C2, color='cyan', label = 'C2' )
plt.plot(tspan, C3, color='red', label = 'C3' )
plt.legend(loc = 'best')
plt.show()



```
